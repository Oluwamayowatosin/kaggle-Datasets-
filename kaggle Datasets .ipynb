```python
%matplotlib inline
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
```

```python
#2023 QS World University Rankings
WUR = pd.read_csv("C:\\Datasets\\2023 QS World University Rankings.csv")
WUR
```

```python
WUR.isnull().sum()
```

```python
WUR = WUR.drop(labels=["location code"], axis=1)
WUR.head()
```

```python
WUR = WUR.set_index(WUR['Rank']).drop(['Rank'], axis=1)
WUR.head()
```

```python
WUR['er score'].fillna(method='ffill', inplace=True)
WUR['fsr score'].fillna(method='ffill', inplace=True)
WUR['cpf score'].fillna(method='ffill', inplace=True)
WUR['ifr score'].fillna(method='ffill', inplace=True)
WUR['ifr rank'].fillna(method='ffill', inplace=True)
WUR['isr score'].fillna(method='ffill', inplace=True)
WUR['isr rank'].fillna(method='ffill', inplace=True)
WUR['irn score'].fillna(method='ffill', inplace=True)
WUR['irn rank'].fillna(method='ffill', inplace=True)
WUR['ger score'].fillna(method='ffill', inplace=True)
WUR['ger rank'].fillna(method='ffill', inplace=True)
WUR.head()
```

```python
WUR.isnull().sum()
```

```python
figure = px.sunburst(WUR, path=["location", "institution"], values="ger score")
figure.show()
```

# Turkey Earthquakes 

```python
Earthquakes  = pd.read_csv("C:\\Datasets\\all_month.csv")
Earthquakes.head()
```

```python
Earthquakes.isnull().sum()
```

```python
n = Earthquakes['nst'].mean()
g = Earthquakes['gap'].mean()
d = Earthquakes['dmin'].mean()
h = Earthquakes['horizontalError'].mean()
mE = Earthquakes['magError'].mean()
m = Earthquakes['magNst'].mean()

print(n)
print(g)
print(d)
print(h)
print(mE)
print(m)
```

```python
Earthquakes['nst'].fillna(n, inplace=True)
Earthquakes['gap'].fillna(g, inplace=True)
Earthquakes['dmin'].fillna(d, inplace=True)
Earthquakes['horizontalError'].fillna(h, inplace=True)
Earthquakes['magError'].fillna(mE, inplace=True)
Earthquakes['magNst'].fillna(m, inplace=True)
Earthquakes.isnull().sum()
```

```python
Earthquakes = Earthquakes.drop(labels=["locationSource", "magSource", "magType", "net"], axis=1)
Earthquakes.head()
```

```python
Earthquakes.plot(kind='scatter', x='longitude', y='latitude', alpha=0.4, s=Earthquakes['gap']/100, label='gap',
figsize=(12, 8), c='magNst', cmap=plt.get_cmap('jet'), colorbar=True)
plt.legend();
```
